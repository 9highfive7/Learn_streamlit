```python
days = 20
tickers ={
    'apple': 'AAPL'
    'facebook':'FB'
}
company = 'apple'

tkr = yf.Ticker(tickers[company])

hist = tkr.history(period=f'{days}d')
hist.index =  hist.index.strftime('%d %B %Y')
hist = hist[['Close']]
hist.columns = [company]
hist = hist.T
hist.index.name = 'Name'
hist
```
